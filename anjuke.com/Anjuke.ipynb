```python
import pandas
df = pandas.read_csv('anjuke.csv')#读取csv文件数据
df.head()
```

# 缺失值处理

```python
#对数据框各列分别进行判断，返回每列是否有缺失值
df.isnull().any()
```

```python
#对数据框df按列统计各列各有多少个缺失值
df.isnull().sum()
```

# 时间日期处理

```python
# 对date列根据'：’：应用切割函数
df['date'] = df['date'].map(lambda e:e.split('：')[1])
```

```python
#将pandas整列数据转换成日期格式
df['date'] = pandas.to_datetime(df['date'],format='%Y年%m月%d日')
df['built'] = pandas.to_datetime(df['built'],format='%Y年')
df.head()
```

```python
#导入时间处理库datetime
from datetime import datetime
#获取系统当前年份
now = datetime.now()
# 计算房龄
df['age'] = now - df['built']
df['age'] = (df['age'].dt.days/365).round(1)
```

```python
df.head()
```

# 标题处理

```python
# 对标题进行处理
df['tittle'] = df['tittle'].map(lambda e:e.strip())
df.head()
```

# 总价处理

```python
# 总价*1000
df['price'] = df['price']*10000
df.head()
```

# 面积处理

```python
# 去面积单位
df['area'] = df['area'].str.extract('([\d+\.\d]*)')
# 转换为float类型
df['area'] = df['area'].astype('float')
df.head()
```

# 计算房屋单价

```python
df['avg'] = (df['price']/df['area']).round(2) # 保留两位小数
df.head()
```
